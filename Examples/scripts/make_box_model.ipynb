```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
from wipy.wipy_utils import utils
from copy import deepcopy
```

```python
model_initl_path = "model_init"

pars = ["vp", "vs", "rho", "x", "z"]

m_init = utils.load_model(model_initl_path, pars)
```

```python
utils.plot_model_fast(m_init, 1, "vs")
```

```python
m_pert = deepcopy(m_init)

bol = (m_pert["x"] > 80.0) * (m_pert["x"] < 120.0) * (m_pert["z"] > 2450.0) * (m_pert["z"] < 2470.0) 

m_pert["vp"] += 1.1*bol*m_pert["vp"]
m_pert["vp"][m_pert["vp"] > 5000.0] = 5000.0
m_pert["vs"] += 1.1* bol* m_pert["vs"]
m_pert["vs"][m_pert["vs"] > 3000.0] = 3000.0

utils.plot_model_fast(m_pert, 1, "vp")
```

```python
output_path = "box_model"

utils.write_model(output_path, m_pert)
```
